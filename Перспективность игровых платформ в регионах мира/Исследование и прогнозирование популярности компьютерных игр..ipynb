В столбце 'user_score' обнаружилось значение 'tbd', вероятнее всего это просто опечетка. Такие строки придется заменить на NaN, так как их доля в датасете велика и подобрать к нему соответствующее числовое значение не получится.

### Изменение динамики продаж по платформам

Кол-во продаж на наиболее популярных платформах.

```python
din_sales = data.pivot_table(index='platform',values='total_sales',
                             aggfunc='sum').sort_values(by='total_sales',ascending=False)#.head(7)
year_din_sales =  data.pivot_table(index='platform',columns='year_of_release',values='total_sales',
                             aggfunc='sum', margins=True, margins_name= 'total_sales')
year_din_sales = year_din_sales.drop('total_sales')
year_din_sales1 = year_din_sales.sort_values(by='total_sales',ascending=False).head(7)
display(year_din_sales1)
din_sales.columns = ['sum_sales']
din_sales
```

По первой таблице можно отследить кол-во выпускаемых игр на платформах по годам. По ней можно отследить выбросы данных, когда определенная игра выходила раньше остальных на платформе и имеет ограниченное число продаж, но она не отражает те строки, где 'year_of_release'==NaN, поэтому для поиска наиболее популярных воспользуемся второй таблицей.

Для некоторых платформ выпуски игр были единичны, возмем для расчета графика продаж и времени появления и исчезновения платформ 7 наиболее популярных. Установим года выпуска первых и последних игр платформы.

```python
col = din_sales.index.to_list()
data_col = data.query('platform in @col')
year_work_platform = data_col.pivot_table(index='platform',values='year_of_release',aggfunc=['min','max']).reset_index()
year_work_platform.columns = ['platform','min_year','max_year']
#year_pl = data_col.groupby('platform')['total_users'].max()
#year = data.
display(year_work_platform)
reg_platform = data_col.pivot_table(index='platform', columns= 'year_of_release',
                        values='total_sales', aggfunc='sum').T.plot(title='Кол-во ежегодных продаж',figsize=(12,12),grid=True);
#reg_platform
```

У платформы DS есть выброс это игра с 1985 годом выпуска, кол-во выпущенных ее копий незначительно, поэтому надо убрать ее из исследования, так как результаты сравнения времени жизни этой платформы могут сильно исказить общий результат. Также можно сказать о платформе X360, у которой есть одна единственная игра старше 2005 года.

```python
display(data[(data['year_of_release']<=2000) & (data['platform']=='DS')])
ds_query = data[(data['year_of_release'] >1985) &(data['platform']=="DS")]
ds_min_year = ds_query['year_of_release'].min()

display(data[(data['year_of_release']<2005) & (data['platform']=='X360')])
x_query = data[(data['year_of_release'] >2000) &(data['platform']=="X360")]
x_min_year = ds_query['year_of_release'].min()
```

Создаем таблицу с пиковыми годами по кол-ву продаж и объединяем ее с таблицей дат выпуска первых и последних игр платформы.

```python

sales_by_year = data_col.pivot_table(index=['platform','year_of_release'],
                                     values='total_sales',aggfunc='sum').reset_index()
max_sales = sales_by_year[sales_by_year['total_sales'] == sales_by_year.groupby(['platform']
                                                                    )['total_sales'].transform(max)].reset_index()
max_sales.columns = ['index','platform','year_max_sales','total_sales']
year_working = pd.concat([year_work_platform,max_sales['year_max_sales']],axis=1,ignore_index=False)
year_working.info()

year_working.loc[year_working['platform']=='XOne','min_year'] =x_min_year 
year_working.loc[year_working['platform']=='DS','min_year'] =ds_min_year 
#year_working['sales_growth'] = year_working['year_max_sales'] - year_working['min_year']
#year_working['falling_sales'] = year_working['max_year'] - year_working['year_max_sales']
year_working
```

### Актуальный период

За величину расчета актуального периода возьмем данные за последние 3 года, иначе есть вероятность захватить данные по устаревшим платформам.

```python
year_working['life_cycle'] = year_working['max_year'] - year_working['min_year']
display(year_working)
life_cycle = 3
#life_cycle = math.floor(year_working['life_cycle'].mean())
#print(f'Средний жизненый цикл платформ == {life_cycle}')
```

<b>При построении дальнейших прогнозов будем ориентироваться на данные, о выпущенных играх, за последние 3 года.</b>

### Выборка актуальных данных

```python
games = data[data['year_of_release']>=(2016-life_cycle)]
games.head(10)
```

```python
act_sales = games.pivot_table(index='platform',values='total_sales',
                             aggfunc='sum').sort_values(by='total_sales',ascending=False).head(7)
act_sales
```

```python
popul_platform = act_sales.index.to_list()
new_platform = games.query('platform in @popul_platform')
profit_platform = new_platform.pivot_table(index='platform', columns= 'year_of_release',
                                        values='total_sales', aggfunc='sum').T.plot(title='Сумма ежегодных продаж',grid=True);
```

Как было заявлено в задании данные за 2016 год могут быть неполными, поэтому стремление графика к 0 в 2016 году можно проигнорировать, а так как продажи упали в целом после 2008 года, стоит рассмотреть полученый выше график более детально и выявить наиболее стабильные платформы.

```python
reg_platform = new_platform[(new_platform['year_of_release']>2008) &
                        (new_platform['year_of_release']<2016)].pivot_table(index='platform', columns= 'year_of_release',
                        values='total_sales', aggfunc='sum').T.plot(title='Сумма ежегодных продаж',grid=True);
```

<b>Резкий подъем продаж игр пока демонстрирует только платформа PS4, в виду своей новизны. Наиболее стабильный доход у платформы 3DS. Также несмотря на свое постепенное угасаное, пока нельзя сбрасывать со счетов платформы X360 и PS3, возможно лишняя рекламная компания им поможет. Все выше перечисленные платформы пока оставим в списке наиболее прибылльных на данный момент и сформируем по ним графики продаж.</b>

## Проверка гипотез

### Сравнение пользовательских рейтингов на платформах PC и XOne

<b>H(0) - средние пользовательские рейтинги на платформах одинаковые.  
    Н(1) - средние пользовательские рейтинги разные.</b>

Сформируем срезы данных по платформам XOne и PC из генеральной совокупности. Сравним их средние значения Ттестом st.ttest_ind.

```python

mean_pc_user_score = data[(data['platform'] == 'PC') & (
    data['user_score'].notna())]['user_score']
mean_xone_user_score = data[(data['platform'] == 'XOne') & (
    data['user_score'].notna())]['user_score']
alpha = .05
result = st.ttest_ind(mean_pc_user_score, mean_xone_user_score, equal_var = False)
print(data[(data['platform'] == 'PC') & (
    data['user_score'].notna())]['user_score'].mean() -
    data[(data['platform'] == 'XOne') & (
    data['user_score'].notna())]['user_score'].mean())
print('Кол-во игр, выпущенных на платформах:',
      len(mean_pc_user_score),'-',len(mean_xone_user_score))
print('p-значение:', result.pvalue)
if result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<b> Полученное число p-value говорит о том, что средние значения пользовательских рейтингов из выборок не одинаковы. А равенство таких значений возможно с вероятностью 2,15 * 10^(-7)%. Такая вероятность слишком маленькая, что говорит о большой разнице между значениями выборок, также на полученную разницу могло повлиять и кол-во выпущенных игр на платформах, ведь у платформы с большИм кол-вом игр вероятность того что какая-нибудь игра стрельнет высоким рейтингом будет выше.</b>

### Сравнение пользовательских рейтингов в жанрах Action и Sports

<b> H(0) - пользовательские рейтинги в жанрах Action и Sports одинаковые.  
    Н(1) - пользовательские рейтинги жанров разнятся.</b>

```python
mean_action_user_score = data[(data['genre'] == 'Action') & (
    data['user_score'].notna())]['user_score']
mean_sports_user_score = data[(data['genre'] == 'Sports') & (
    data['user_score'].notna())]['user_score']
alpha = .05
result = st.ttest_ind(mean_action_user_score, mean_sports_user_score, equal_var = False)
print(data[(data['genre'] == 'Action') & (
    data['user_score'].notna())]['user_score'].mean() -
    data[(data['genre'] == 'Sports') & (
    data['user_score'].notna())]['user_score'].mean())
print('Кол-во игр, выпущенных на платформах:',
      len(mean_action_user_score),'-',len(mean_sports_user_score))
print('p-значение:', result.pvalue)
if result.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<b> Полученое число p-value говорит о том, что пользовательские рейтинги, в сравниваемых жанрах различны. А вероятность с которой случайно возможно получить равенство значений, в пределах допустимой границы значимости, равна 2%. Первая гипотеза определила большую разницу между средними пользовательскими оценками на платформах XOne и PC. Это значит, что значение p-value будет стремиться к 0. P-value меньше допустимой границы значимости, значит вероятность с которой возможно получить равенство оценок пользователей ==p-value *100
